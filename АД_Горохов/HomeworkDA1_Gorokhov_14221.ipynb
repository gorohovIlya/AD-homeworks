**Непрерывные распределения**

**Распределение Ландау**

```python
import scipy.stats as ss
N = 10000
# Распределение Ландау
x = ss.landau()
d = x.rvs(N)
# Гистограмма
import matplotlib.pyplot as plt
plt.hist(d, bins=100)
```

```python
# Функция распределения
plt.hist(d, cumulative=True, density=True, bins=100)
```

```python
# Анализ ошибки определения среднего и СКО по N случайным значениям из распределения и его приближения по центральной предельной теореме в зависимости от N
import numpy as np
sample_sizes = [10, 100, 1000, 10000] # Различные размеры выборок
errors_mean = []
errors_std = []
for sz in sample_sizes:

  sample = ss.landau().rvs(sz)
  sample_std = sample.std()
  sample_mean = sample.mean()
  normal = ss.norm(sample_mean, sample_std).rvs(sz)
  sample_mean = sample.mean()
  normal_mean = normal.mean()
  sample_std = sample.std()
  normal_std = normal.std()
  errors_mean.append(abs(sample_mean - normal_mean))
  errors_std.append(abs(sample_std - normal_std))
  print(f"for size {sz} absolute mean value error: {abs(sample_mean - normal_mean)}, absolute std error: {abs(sample_std - normal_std)}")
fig, ax = plt.subplots(1, 2)
ax[0].set_ylim(0, max(errors_mean))
ax[0].set_title("MEAN ERRORS")
ax[0].plot(errors_mean)
ax[1].set_ylim(0, max(errors_std))
ax[1].set_title("STD ERRORS")
ax[1].plot(errors_std)
```

Сравнение гистограммы распределения Ландау и нормального распределения

```python
u = ss.landau().rvs(10000)
mu_x,std_x= u.mean(),u.std()
n1 = ss.norm(mu_x, std_x).rvs(10000)
plt.hist(u,bins=100,density=True,label='landau');
plt.hist(n1,bins=100,density=True,histtype='step',label='norm');
plt.legend()
```
