```python
f = open("data.txt")
a, b, c, d = 0, 0, 0, 0
for s in f:
  s = s.split()
  if s[1] == 'Yes' and s[2] == 'Yes':
    d += 1
  elif s[1] == 'Yes' and s[2] == 'No':
    b += 1
  elif s[1] == 'No' and s[2] == 'Yes':
    c += 1
  else:
    a += 1
print(a, b, c, d)
```

```python
import scipy.stats as ss
import numpy as np
n = a + b + c + d
z = (b - c) / np.sqrt((b + c) - (((b - c) ** 2) / n))
pv = 2 * (1 - ss.norm(0, 1).cdf(np.abs(z)))
if pv > 0.05:
  print(f"p-value = {pv:.2}, {pv:.2} > 0.05. Н0 не может быть отвергнута -> общественное мнение не претерпело значительных изменений")
else:
  print(f"p-value = {pv:.2}, {pv:.2} <= 0.05. Отвергаем Н0 -> общественное мнение изменилось значительно")
```
