```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

# Непрерывные распределения

### Распределение Вальда (обратное Гауссово)

```python
N = 10000# размер выборки
x = st.wald.rvs(size = N)
plt.hist(x, bins = 100)
plt.title("Распределение Вальда")
plt.show()
```

```python
N = 10000
x = st.wald.rvs(size = N)
plt.hist(x, bins = 100, cumulative = True, density = True)
plt.title("Функция распределения")
plt.show()
```

```python
mean_x = []
std_x = []
for _ in range(500):
    x = st.wald.rvs(size = N)
    mean_x.append(abs(st.wald.mean() - np.mean(x)))
    std_x.append(abs(st.wald.std() - np.std(x, ddof = 1)))
    
print(f"Ошибка определения среднего при N = {N}:\t",np.mean(mean_x))
print(f"Ошибка определения СКО при N = {N}:\t", np.mean(std_x))
```

```python
N = 100_000 # увеличим N
mean_x = []
std_x = []
for _ in range(500):
    x = st.wald.rvs(size = N)
    mean_x.append(abs(st.wald.mean() - np.mean(x)))
    std_x.append(abs(st.wald.std() - np.std(x, ddof = 1)))
    
print(f"Ошибка определения среднего при N = {N}:\t",np.mean(mean_x))
print(f"Ошибка определения СКО при N = {N}:\t", np.mean(std_x))
```

```python
N = 1_000_000 # увеличим N еще раз
mean_x = []
std_x = []
for _ in range(500):
    x = st.wald.rvs(size = N)
    mean_x.append(abs(st.wald.mean() - np.mean(x)))
    std_x.append(abs(st.wald.std() - np.std(x, ddof = 1)))
    
print(f"Ошибка определения среднего при N = {N}:\t",np.mean(mean_x))
print(f"Ошибка определения СКО при N = {N}:\t", np.mean(std_x))
```

Как можно заметить, с увеличением выборки уменьшается ошибка определения среднего и СКО распределения, что подтверждает ЦПТ. 

Для наглядности построим графики зависимости ошибки от выборки

```python
Ns = [10,100,1000,10_000,100_000,1_000_000]

mean_err = []
std_err = []

for N in Ns:
    mean_x = []
    std_x = []
    for _ in range(500):
        x = st.wald.rvs(size = N)
        mean_x.append(abs(st.wald.mean() - np.mean(x)))
        std_x.append(abs(st.wald.std() - np.std(x, ddof = 1)))
        
    mean_err.append(np.mean(mean_x))
    std_err.append(np.mean(std_x))

plt.figure(figsize=(8,5))
plt.loglog(Ns, mean_err, 'o-', label="Ошибка среднего")
plt.loglog(Ns, std_err, 's-', label="Ошибка СКО")
plt.title("Зависимость ошибки от N (распределение Вальда)")
plt.xlabel("Размер выборки")
plt.ylabel("Величина ошибки")
plt.show()
```

На графике видно подтверждение ЦПТ. Проверим ее и на других распределениях
