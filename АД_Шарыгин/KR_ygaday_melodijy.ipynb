```python
import scipy.stats as ss
import numpy as np
```

```python
p = 89 / 100
p0 = 0.85
N = 100

Z = (p - p0) / (np.sqrt(p0 * (1 - p0) / N))

d = ss.norm(0, 1)
xL = d.ppf(0.025)
xH = d.ppf(0.975)

print(f'Z = {Z}')
print(f'xL = {xL}')
print(f'xH = {xH}')

if xL <= Z <= xH:
    print("моя нейронка круче")
else:
    print("моя нейронка хуже()")
```
