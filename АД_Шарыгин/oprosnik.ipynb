```python
import numpy as np
import scipy.stats as ss
```

```python
vote1_list = []
vote2_list = []
with open('data.txt', 'r', encoding='utf-8') as file:
    for line in file:
        line = line.strip()
        if line:
            parts = line.split('\t')
            if len(parts) >= 3:
                vote1_list.append(parts[1].strip())
                vote2_list.append(parts[2].strip())
```

```python
vote1 = np.array(vote1_list)
vote2 = np.array(vote2_list)
```

```python
b = np.sum((vote1 == 'No') & (vote2 == 'Yes'))
c = np.sum((vote1 == 'Yes') & (vote2 == 'No'))
```

```python
Z = (b - c) / np.sqrt(b + c - (b - c)**2 / len(vote1))
pval = 2 * (1 - (ss.norm(0, 1).cdf(abs(Z))))
print(f'{Z=}, {pval=}')
```

```python
if pval < 0.05:
    print('Мнение изменилось')
else:
    print('Мнение не изменилось')
```
